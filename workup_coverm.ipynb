```python
import glob

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
import os.path as op
```

```python
tsvs = glob.glob('/mnt/stepanauskas_nfs/projects/sag_mag/coverm/1Msub/*concat_coverm-exclude_secondary*.tsv')
```

```python
tsv = tsvs[0]

gtypes = []
mgs = []
counts = []

for tsv in tsvs:
    try:
        df = pd.read_csv(tsv, sep = "\t")
        df.columns = [i.split(" ")[-1] for i in df.columns]
        counts.append(df['Count'].sum())
        gtypes.append(op.basename(tsv).split("_")[0])
        mgs.append(op.basename(tsv.split("_")[-1].replace('.tsv','')))

    except:
        continue
```

```python
rdf = pd.DataFrame(data = {'mg':mgs, 'genome_type':gtypes, 'reads_recruited':counts})
```

```python
rdf['pct_recruited'] = rdf['reads_recruited'] / 1000000 * 100
rdf
```
